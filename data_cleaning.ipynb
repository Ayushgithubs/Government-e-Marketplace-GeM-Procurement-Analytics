```python
df["award_date"] = pd.to_datetime(
    df["award_date"],
    format="%d-%m-%Y"
)
```

```python
df["award_date"].dtype
```

```python
df["award_year"] = df["award_date"].dt.year
```

```python
df["award_month"] = df["award_date"].dt.month
```

```python
df["award_month_name"] = df["award_date"].dt.month_name()
```

```python
df["award_quarter"] = df["award_date"].dt.quarter
```

```python
df["award_day_name"] = df["award_date"].dt.day_name()
```

```python
df.to_csv("government_procurement_cleaned.csv", index=False)
```
